### 线性变换的PCA 和MDS

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import MDS
clf2 = MDS(2)
clf2.fit(iris.data)
iris_t2 = clf2.fit_transform(iris.data)
plt.scatter(iris_t2[:,0],iris_t2[:,1],c=iris.target)
plt.title('using sklearn MDS')
plt.show()
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
clf2.fit(digits.data)
digits_t2 = clf2.fit_transform(digits.data)
plt.scatter(digits_t2[:,0],digits_t2[:,1],c=digits.target)
plt.title('using sklearn MDS')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
digits_pca = pca.fit_transform(digits.data)
plt.scatter(digits_pca[:,0],digits_pca[:,1],c=digits.target
           ,alpha=.5)
plt.colorbar()
plt.title('Using sklearn PCA')
plt.show()
```

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
digits_iso = iso.fit_transform(digits.data)
plt.scatter(digits_iso[:,0],digits_iso[:,1],c=digits.target,alpha=0.5)
plt.colorbar()
plt.title('Using sklearn Isomap')
plt.show()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=3,init='pca')
digits_tsne = tsne.fit_transform(digits.data)

plt.scatter(digits_tsne[:,0],digits_tsne[:,1],c=digits.target,alpha=0.5)
plt.colorbar()
plt.title('Using sklearn TSNE')
plt.show()
```
